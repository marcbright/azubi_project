```python
numerical_cols = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
```

```python
# Set up the plotting environment
plt.figure(figsize=(15, 10))

# Box plots to identify outliers
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=bank_full[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Histogram to visualize distribution
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.histplot(bank_full[col], kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Calculate IQR and identify outliers
for col in numerical_cols:
    Q1 = bank_full[col].quantile(0.25)
    Q3 = bank_full[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = bank_full[(bank_full[col] < lower_bound) | (bank_full[col] > upper_bound)][col]
    print(f"\nOutliers in {col}:")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"Number of outliers: {len(outliers)}")
    print(f"Outlier values: {outliers.head() if not outliers.empty else 'None'}")
```

```python
# Cap outliers for balance
Q1_balance = bank_full['balance'].quantile(0.25)
Q3_balance = bank_full['balance'].quantile(0.75)
IQR_balance = Q3_balance - Q1_balance
lower_bound_balance = Q1_balance - 1.5 * IQR_balance
upper_bound_balance = Q3_balance + 1.5 * IQR_balance
bank_full['balance'] = bank_full['balance'].clip(lower=lower_bound_balance, upper=upper_bound_balance)
```

```python
# Cap outliers for duration
Q1_duration = bank_full['duration'].quantile(0.25)
Q3_duration = bank_full['duration'].quantile(0.75)
IQR_duration = Q3_duration - Q1_duration
upper_bound_duration = Q3_duration + 1.5 * IQR_duration
bank_full['duration'] = bank_full['duration'].clip(upper=upper_bound_duration)
```

```python
# Cap outliers for campaign
Q1_campaign = bank_full['campaign'].quantile(0.25)
Q3_campaign = bank_full['campaign'].quantile(0.75)
IQR_campaign = Q3_campaign - Q1_campaign
upper_bound_campaign = Q3_campaign + 1.5 * IQR_campaign
bank_full['campaign'] = bank_full['campaign'].clip(upper=upper_bound_campaign)
```

```python
# Custom handling for pdays (cap at 365 days)
bank_full['pdays'] = bank_full['pdays'].clip(upper=365)

# Custom handling for previous (cap at 5)
bank_full['previous'] = bank_full['previous'].clip(upper=5)
```

```python
print("\nSummary after outlier handling:")
bank_full[numerical_cols].describe()
```

```python
# Encode y as binary (0 = no, 1 = yes)
bank_full['y'] = bank_full['y'].cat.codes
```

```python
bank_full["y"].unique()
```

```python
# Transform pdays: binarize (-1 vs. positive)
bank_full['pdays_binarized'] = (bank_full['pdays'] > -1).astype('int')
```

```python
# Recalculate crosstab with transformed pdays
print("\nCross-tabulation of pdays_binarized and y:")
pd.crosstab(bank_full['pdays_binarized'], bank_full['y'])
```

```python
# Recalculate correlation matrix with pdays_binarized
numerical_cols = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays_binarized', 'previous']
correlation_matrix = bank_full[numerical_cols + ['y']].corr()
```

```python
# Plot updated correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix with Transformed pdays')
plt.show()
```

```python
# Patterns in categorical variables (bar plots)
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
plt.figure(figsize=(20, 15))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=bank_full, x=col, hue='y')
    plt.xticks(rotation=40)
    plt.title(f'{col} vs y')

plt.tight_layout()
plt.show()
```

```python
# Summary of y distribution
print("Distribution of y:")
bank_full['y'].value_counts(normalize=True)
```

```python
# One-hot encode categorical variables
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_cats = encoder.fit_transform(bank_full[categorical_cols])
encoded_df = pd.DataFrame(encoded_cats, columns=encoder.get_feature_names_out(categorical_cols))
bank_full_encoded = pd.concat([bank_full[numerical_cols + ['y']], encoded_df], axis=1)
```

```python
# Create interaction features
bank_full_encoded['duration_campaign_interaction'] = bank_full_encoded['duration'] * (1 / (bank_full_encoded['campaign'] + 1))  # Inverse campaign effect
bank_full_encoded['pdays_prev_interaction'] = bank_full_encoded['pdays_binarized'] * bank_full_encoded['previous']
```

```python
# Separate features and target
X = bank_full_encoded.drop('y', axis=1)
y = bank_full_encoded['y']
```

```python
# Handle imbalance with SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Verify class balance
print("Class distribution before SMOTE:")
print(y.value_counts(normalize=True))
print("\nClass distribution after SMOTE:")
print(pd.Series(y_resampled).value_counts(normalize=True))
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

print("\nShape of training set:", X_train.shape)
print("Shape of test set:", X_test.shape)
```

```python
# Initialize models
log_reg = LogisticRegression(random_state=42, max_iter=1000)
rf_clf = RandomForestClassifier(random_state=42, n_estimators=100)
```

```python
# Train models
log_reg.fit(X_train, y_train)
rf_clf.fit(X_train, y_train)
```

```python
# Predict
y_pred_log_reg = log_reg.predict(X_test)
y_pred_rf = rf_clf.predict(X_test)
```

```python
# Evaluate logistic regression
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_log_reg))
print("\nLogistic Regression Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_log_reg))
```

```python
# Evaluate random forest
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
print("\nRandom Forest Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
```

```python
# Define parameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Initialize Random Forest
rf_clf = RandomForestClassifier(random_state=42)

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=rf_clf, param_grid=param_grid, cv=5, scoring='f1_macro', n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
# Best parameters and score
print("Best parameters:", grid_search.best_params_)
print("Best cross-validation F1 score:", grid_search.best_score_)
```
